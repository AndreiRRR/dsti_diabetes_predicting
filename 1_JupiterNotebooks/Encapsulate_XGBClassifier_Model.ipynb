Objectif : create the pickle archive of the best prediction model ready to use in Web Diabetes Predicting Application. 

XGBClassifier model shows the best results for diabetes prediction (see Diabetes_Predicting notebook).
This notebook trains XGBClassifier model, tests it and creates a pickle archive which we can reuse in Web Application.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
from xgboost import XGBClassifier
```

```python
data = pd.read_csv("Data\\TAIPEI_diabetes.csv")
```

```python
data.head()
```

```python
X = data.drop(columns=["Diabetic"])
y = data["Diabetic"]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

```python
model = XGBClassifier(eval_metric="logloss", random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("accuracy: ", round(accuracy,2)) 
print("accuracy: ", accuracy) 
```

```python
y_pred = model.predict(X_train)
accuracy = accuracy_score(y_train, y_pred)
print("accuracy: ", round(accuracy,2)) 
print("accuracy: ", accuracy) 
```

```python
with open("Diabetes_Model.pkl", "wb") as model_file : 
    pickle.dump(model, model_file)
```
